````python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
````

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

````python
class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        np.random.seed(self.random_state)
        self.theta =0
        self.list_of_costs=[]


    def sigmoid(self, X):
        return (1 /(1 + np.exp(-np.matmul(X, self.theta))))

    
    def cost(self, X ,y):
        m = X.shape[0]
        return ((-(np.matmul(y, np.log(self.sigmoid(X)))) - (np.matmul(1 - y, np.log(1-self.sigmoid(X))))) / m)
    
    def cost_derivative (self, X, y):
        return np.matmul(self.sigmoid(X)-y, X)
    
    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        # generate random theta
        self.theta = np.random.random( X.shape[1])
        self.list_of_costs.append(self.cost(X,y))
        improve=1+self.eps
        i=1;
        
        while(i<self.n_iter and self.eps<=improve):
            self.theta -= (self.eta)*self.cost_derivative(X,y)
            self.list_of_costs.append(self.cost(X,y))
            improve=self.list_of_costs[i-1]-self.list_of_costs[i]
            i+=1
        
                      
  
    def predict(self, X):
        """Return the predicted class label"""
        if (X.shape[1] == 2):
            X = np.insert(X,0,1,axis=1)
        return np.rint(self.sigmoid(X))
              
````

## Cross Validation

(5 points)

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

Shuffle the training set before you split the data to the folds.

````python
### Your code here ####
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]

training_set_shuffeld = np.insert(training_set.values,0,1,axis=1)
np.random.seed(1)
np.random.shuffle(training_set_shuffeld)
fold_k = 5
fold_length = (int)(training_set_shuffeld.shape[0] / fold_k)
accuracy_dict = {}

def split_data_val_train(i, train_set, fold_k, fold_length):
    cur_val = train_set[i * fold_length : (i+1) * fold_length,:]
    cur_train = np.delete(train_set, list(range(i * fold_length, (i+1) * fold_length)), 0)
    val_X = cur_val[:,:-1]
    val_y =cur_val[:,-1]
    train_X = cur_train[:,:-1]
    train_y = cur_train[:,-1]
    return val_X, val_y, train_X, train_y

def accuracy(predictions, labels, k):
    success = 0.0
    for i in range(len(labels)):
        if predictions[i] == labels[i]:
            success += 1
    return 100*(success / (labels.shape[0]))  
    

for i in range(fold_k):
    val_X, val_y, train_X, train_y = split_data_val_train(i, training_set_shuffeld, fold_k, fold_length)
    
    for eta in etas:
        for eps in epss: 
            cur_logistic_rigression = LogisticRegressionGD(eta, 10000, eps)
            cur_logistic_rigression.fit(train_X, train_y)
            predicts = cur_logistic_rigression.predict(val_X)
            cur_accuracy = accuracy(predicts, val_y,fold_k)
            sum_of_accuracies = accuracy_dict.get((eta, eps), 0)
            accuracy_dict[(eta,eps)] = (sum_of_accuracies + (cur_accuracy)/fold_k)
            
    
best_eta_and_eps = max(accuracy_dict, key=accuracy_dict.get)

print ("Best ETA: " , best_eta_and_eps[0], "\nBest EPS: ", best_eta_and_eps[1], "\nWith average accuracy of: ", max(accuracy_dict.values()), "%")

best_eta, best_eps = best_eta_and_eps

````

## Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

````python
#### Your code here ####
# calc normal pdf    
def norm_pdf(data, mu, sigma):
    normal_distribution=(1/(2*np.pi*(sigma**2))**0.5)*np.e**((-(data-mu)**2)/(2*(sigma**2)))
    return normal_distribution
````

## Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

````python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1, n_iter=1000, eps=0.01, random_state=1):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.mus = []
        self.sigmas = []
        self.weights = []
        self.n_samples = 0
        self.J_history = []
        
        

    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        """
        self.data = data
        self.n_samples = self.data.shape[0]
        np.random.seed(self.random_state)
        self.mus = np.zeros(self.k)
        self.sigmas = np.zeros(self.k)
        self.weights = np.ones(self.k)/self.k
        splited_data = np.array(np.array_split(data, self.k))
        for i in range(self.k):
            self.mus[i] = np.mean(splited_data[i])
            self.sigmas[i] = np.std(splited_data[i])
        
        
       
    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        responsibilities = np.zeros((len(data),1))
        denominator = np.zeros((len(data),1))
        
        for i in range(self.k):
            numerator = self.weights[i]* norm_pdf(data,self.mus[i],self.sigmas[i])
            responsibilities = np.concatenate((responsibilities, numerator),axis=1)
            denominator = np.add(numerator, denominator)
        
        responsibilities = responsibilities[:,1:self.k+1]
        responsibilities = responsibilities / denominator
        self.responsibilities = responsibilities

        

    def maximization(self, data):
        """
        M step - updating distribution params
        """
        self.n_samples = len(data)
        for i in range(self.k):
            self.weights[i] = np.sum(self.responsibilities[:,i], axis = 0) / self.n_samples
            reshaped_responsibility_culomn = np.reshape(self.responsibilities[:,i], (self.n_samples,1)) 
            self.mus[i] = (1.0 / (self.weights[i] * self.n_samples)) *  np.sum(reshaped_responsibility_culomn * data , axis = 0)
            self.sigmas[i] = ((1.0 / (self.weights[i] * self.n_samples)) * np.sum(reshaped_responsibility_culomn * (data - self.mus[i])**2))**0.5

        
        
    def cost_log(self, data, weight, mu, sigma):
        pdf = norm_pdf(data, mu, sigma)  
        return np.sum(-np.log(weight * pdf))
    
    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then expectation and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.
        """
        self.init_params(data)
        
        
        J_current = -1
        J_previous = 0
        prev_cost = 0
        
        
        for i in range(self.n_iter):
            
            #E step
            self.expectation(data)
            
            #M step
            self.maximization(data)
            
            J_previous = J_current
      
            J_current = 0
            for j in range(self.k):
                J_current += self.cost_log(data, self.weights[j], self.mus[j], self.sigmas[j])
                
            self.J_history.append(J_current)
            
            if ((J_previous != -1) and (np.abs(J_current - J_previous) < self.eps)):
                break
                           
            
        #plt.plot(np.arange(self.n_iter), self.J_history)
        #plt.xlabel('Iterations')
        #plt.ylabel('Loss')
        #plt.title('Loss as a function of iterations')
        #plt.show()         
        #    
    def get_dist_params(self):
        return self.weights, self.mus, self.sigmas
````

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


````python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1):
        self.k = k
        self.prior = 0
        self.weights = []
        self.mus = []
        self.sigmas = []
        self.num_of_classes = 0
        self.num_of_features = 0
        

    def fit(self, X, y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        
        #Calculate prior
        uniques , counts = np.unique(y , return_counts = True)
        self.prior = (counts / len(y))
        self.num_of_classes = len(uniques)
        self.num_of_features = X.shape[1]
        
        
        self.weights = np.zeros([self.num_of_classes, self.num_of_features, self.k])
        self.mus = np.zeros([self.num_of_classes, self.num_of_features, self.k])
        self.sigmas = np.zeros([self.num_of_classes, self.num_of_features, self.k])
        
        for i in range(self.num_of_classes):
            data_of_class = X[np.where(y == uniques[i])]
            for j in range(X.shape[1]):
                feature_of_class = data_of_class[:, j:j+1]
                cur_class_and_feature_em = EM(self.k)
                cur_class_and_feature_em.fit(feature_of_class)
                self.weights[i][j], self.mus[i][j], self.sigmas[i][j] = cur_class_and_feature_em.get_dist_params()
                
#         print(self.weights)
#         print(self.mus)
#         print(self.sigmas)
        
    
    def predict(self, X):
        """Return the predicted class label"""
        prediction = np.zeros([self.num_of_classes, len(X)])
        
        for label in range(self.num_of_classes):
            cur_prior = self.prior[label]
            cur_likelihood = 1
            for feature in range(self.num_of_features):
                density_function = 0
                for k in range(self.k):
                    pdf = norm_pdf(X[:,feature], self.mus[label][feature][k], self.sigmas[label][feature][k])
                    density_function += self.weights[label][feature][k] * pdf
                cur_likelihood *= density_function
            prediction[label] =  cur_prior*cur_likelihood
        return np.argmax(prediction, axis=0)
````

````python
from scipy.stats import multivariate_normal

pi = np.array([1000, 1000, 1000,1000])
mu = [[10, 5, 0], [-10, -5, 0], [-10, 5, 3],[10, -5, -3]]
sigma = [ [[9, 0, 0], 
           [0, 9, 0],
           [0, 0, 9]],
         
          [[9, 0, 0], 
           [0, 9, 0],
           [0, 0, 9]],         
         
          [[9, 0, 0], 
           [0, 9, 0],
           [0, 0, 9]],  
         
          [[9, 0, 0], 
           [0, 9, 0],
           [0, 0, 9]]]

#Plot settings
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
color_list=["g","g","b","b"]

#Generate the data
y0 = np.zeros((1000,1)) #class 0
y1 = y0 + 1 #Class 1
class_list=[y0, y0, y1, y1]

x_for_data, y_for_data, z_for_data = (0, 0, 0)
dataset_A = np.empty((0, 4))

#generate 4 normal disterbutions, 2 for each class
for i in range(len(pi)): 
    col=0
    x, y, z = np.random.multivariate_normal(mu[i], sigma[i], pi[i]).T
    x_for_data = np.reshape(x,(x.shape[0],1))
    y_for_data = np.reshape(y,(x.shape[0],1))
    z_for_data = np.reshape(z,(x.shape[0],1))
    data_part = np.concatenate((x_for_data, y_for_data, z_for_data, class_list[i]),axis=1) #Classifier each normal disterbution
    dataset_A = np.concatenate((dataset_A,data_part),axis=0) #adding the features disterbutions to the dataset
    ax.scatter(x, y, z, c = color_list[i])
    ax.xaxis.set_label_position('top')
    ax.set_xlabel('feature 1')
    ax.set_ylabel('feature 2')
    ax.set_zlabel('feature 3')
    ax.set_title('Naive Bayes will work better than Logistic Regression!')
    
plt.show()
````

In the above graph the NB works better because the data isn't linear separtable - there isn't separte line between the classes (As we see in the previous question, LG not working so well with this kind of data).
In addition, the feature are linear independent with each other (good for NB).

In the above graph, the data is linear separtable, so we saw in the previous question that both NB and LG can work well with this kind of data.
