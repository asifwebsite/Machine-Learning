```python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

```python
class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.J_history = []
        self.theta = None

    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        np.random.seed(self.random_state)
        
        difference = 0
        J_history = []
        J_current = -1
        J_previous = 0
        
        X = np.column_stack((np.ones(X.shape[0]), X))
        self.theta = np.random.random(size=X.shape[1])
        
        for i in range(self.n_iter):
            hypothesis = self.sigmoid(np.dot(X, self.theta))
            self.theta = self.theta - self.eta * np.dot(X.T, hypothesis - y)
            
            J_previous = J_current
            J_current = self.compute_cost(X, y, self.theta)
            J_history.append(J_current)
            
            difference = np.abs(J_current - J_previous)
            if (J_previous != -1 and  difference < self.eps):
                break
        
        self.J_history = J_history
    
    def predict(self, X):
        """Return the predicted class label"""
        X = np.column_stack((np.ones(X.shape[0]), X))
        predict = np.round(self.sigmoid(np.dot(X, self.theta)))
        
        return predict

    def compute_cost(self, X, y, theta):
        size = len(y)
        h_theta = self.sigmoid(X.dot(theta))
        
        inner_product = (-y).T.dot(np.log(h_theta + self.eps)) - (1 - y).T.dot(np.log(1 - h_theta + self.eps))
        cost = (1 / size) * inner_product
        
        return cost
        
    def sigmoid(self, X):
        
        sigmoid = 1.0 / (1 + np.exp(-X))
        return sigmoid
```

## Cross Validation

(5 points)

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

Shuffle the training set before you split the data to the folds.

```python
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]

num_folds = 5
train_values = training_set.values
train_values_shuffle = train_values.copy()
np.random.shuffle(train_values_shuffle)
x_train_shuffle = train_values_shuffle[:, [0,1]]
y_train_shuffle = train_values_shuffle[:, -1]

X_train_splits = np.array(np.array_split(x_train_shuffle, 5))
y_train_splits = np.array(np.array_split(y_train_shuffle, 5))
max_accuracy = 0

for eta in etas:
    for eps in epss:
        accuracy_sum = 0
        for i in range(num_folds):
            X_train_folds = []
            y_train_folds = []
            for j in range(num_folds):
                if (i != j):
                    X_train_folds.append(X_train_splits[j])
                    y_train_folds.append(y_train_splits[j])
            
            X_train_folds = np.array(X_train_folds)
            y_train_folds = np.array(y_train_folds)
            X_train_folds = X_train_folds.reshape((-1, 2))
            y_train_folds = y_train_folds.ravel()
            validation_fold = [X_train_splits[i], y_train_splits[i]]
            
            logistic = LogisticRegressionGD(eta = eta, eps = eps)
            logistic.fit(X_train_folds, y_train_folds)
            pred = logistic.predict(validation_fold[0])
            val_fold_len = len(validation_fold[1])
            abs_calc = np.abs(pred - validation_fold[1])
            accuracy_sum = accuracy_sum + (1 - (np.sum(abs_calc) / val_fold_len))
            
        if ((accuracy_sum / 5) > max_accuracy):
            max_accuracy = accuracy_sum / 5
            best_eta = eta
            best_eps = eps
```

## Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

```python
# calc normal pdf    
def norm_pdf(data, mu, sigma):
    norm = np.exp(-0.5 * np.square(data - mu) / np.square(sigma)) / np.sqrt(2 * np.pi * np.square(sigma))
    
    return norm
```

## Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

```python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1, n_iter=1000, eps=0.01, random_state = 1, mu=0, sigma=0, weights=None, responsibilties=None, probs = None):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.mu = mu
        self.sigma = sigma
        self.weights = weights
        self.responsibilties = responsibilties
        self.probs = probs

    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        """
        np.random.seed(self.random_state)
        
        self.weights = np.random.random(self.k)
        self.weights = self.weights/np.sum(self.weights)
        
        mus = np.empty(self.k,)
        sigmas = np.empty(self.k,)
        
        data_splits = np.array(np.array_split(data, self.k))
        
        for i in range(self.k):
            mus[i] = np.mean(data_splits[i])
            sigmas[i] = np.std(data_splits[i])
            
        self.mu = mus
        self.sigma = sigmas
        
        
    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        self.probs = np.empty([self.k, data.shape[0]])
        
        for i in range(len(self.probs)):
            self.probs[i] = self.weights[i] * norm_pdf(data, self.mu[i], self.sigma[i])
            
        self.responsibilties = self.probs / np.sum(self.probs, axis = 0)

    def maximization(self, data):
        """
        M step - updating distribution params
        """
        self.weights = np.sum(self.responsibilties, axis = 1) / data.shape[0]
        
        for i in range(self.k):
            self.mu[i] = np.dot(self.responsibilties[i], data) / np.sum(self.responsibilties[i])
            self.sigma[i] = np.sqrt(np.dot(self.responsibilties[i], np.square(data - self.mu[i])) / (np.sum(self.responsibilties[i])))
        

    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then expectation and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.
        """
        
        J_history = []
        J_current = -1
        J_previous = 0
        
        self.init_params(data)
        
        for i in range(self.n_iter):
            self.expectation(data)
            self.maximization(data)
            
            J_previous = J_current
            J_current = self.compute_cost()
            J_history.append(J_current)
            
            if ((J_previous != -1) and (np.abs(J_current - J_previous) < self.eps)):
                break
            
            self.J_history = J_history
            
    
                
    def get_dist_params(self):
        return self.weights, self.mu, self.sigma
    
    def compute_cost(self):
        cost = np.sum(-np.log(np.sum(self.probs, axis = 0)))
        return cost
    
    
    
    
```

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


```python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1,random_state=1 , priors=None, weights=None, mu=None, sigma=None, classes=None):
        self.k = k
        self.random_state = random_state
        self.weights = weights
        self.mu = mu
        self.sigma = sigma
        self.priors = priors
        self.classes = classes

    def fit(self, X, y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        uniques , counts = np.unique(y , return_counts = True)
        self.classes = uniques

        num_of_classes = len(uniques)
        num_of_features = X.shape[1]
        
        self.priors = self.prior(y)
        
        self.weights = np.empty([num_of_classes, num_of_features, self.k])
        self.mu = np.empty([num_of_classes, num_of_features, self.k])
        self.sigma = np.empty([num_of_classes, num_of_features, self.k])
        
        for i in range(num_of_classes):
            data_by_class = X[np.where(y == uniques[i])]
            for j in range(X.shape[1]):
                feature_by_class = data_by_class[:, j]
                class_EM = EM(self.k, random_state = self.random_state)
                class_EM.fit(feature_by_class)
                self.weights[i][j], self.mu[i][j], self.sigma[i][j] = class_EM.get_dist_params()
                
                
    def predict(self, X):
        """Return the predicted class label"""
        num_classes = len(self.classes)
        pred = np.empty([num_classes, len(X)])
        
        for i in range(num_classes):
            prior = self.priors[i]
            likelihood = self.likelihood(X, i)
            pred[i] = prior * likelihood
            
        pred = pred[0] < pred[1]
        pred = pred.astype(int)
        
        return pred
        
    def prior(self, y):
        uniques , counts = np.unique(y , return_counts = True)
        prior_sol = (counts / len(y))
        return prior_sol
    
    def likelihood(self, X, class_value):
        likelihood = 1
        iterations = self.mu.shape[1]
        for i in range(iterations):
            prob = 0
            for j in range(self.k):
                norm = norm_pdf(X[:,i], self.mu[class_value][i][j], self.sigma[class_value][i][j])
                prob = prob + self.weights[class_value][i][j] * norm
            likelihood = likelihood * prob
        return likelihood
        
naive = NaiveBayesGaussian(k=2)
naive.fit(X_training, y_training)
predict = naive.predict(training_set.values)
print(predict)
```

```python
# dataset that Naive Bayes will work better than Logisitc Regression.
samples = 1000
naive_points = []
for i in range(samples):
    param = [[60,4], [6,10]]
    random = np.random.random_sample()
    predict = 1
    if (random < 0.1):
        predict = 0
    mu, sigma = param[predict]
    naive_points.append([np.random.normal(mu, sigma, 1)[0], predict])

# dataset that Logistic Regression will work better than Naive Bayes using EM.
logistic_points = []
theta = np.array([0.5,0.5])
for i in range(samples):
    random = np.random.random_sample()*4-2
    hypotesis = 1 / (1 + np.exp(-np.dot(random, theta.T)))[0]
    predict = 0
    if (hypotesis > 0.5):
        predict = 1
    logistic_points.append([random,predict])
    
# fill arrays according to the two classes       
naive_class_0 = []
naive_class_1 = []
y = [x[1] for x in naive_points]
for i in range(len(naive_points)):    
    if (y[i] == 0):
        naive_class_0.append(naive_points[i][0])
    else:
        naive_class_1.append(naive_points[i][0])
        
logistic_class_0 = []
logistic_class_1 = []
y = [x[1] for x in logistic_points]
for i in range(len(logistic_points)):    
    if (y[i] == 0):
        logistic_class_0.append(logistic_points[i][0])
    else:
        logistic_class_1.append(logistic_points[i][0])

# visualizing data
figure, axs = plt.subplots(2, figsize=(8, 8))
axs[0].hist(naive_class_0, bins=20, alpha=0.5, label='class 0')
axs[0].hist(naive_class_1, bins=20, alpha=0.5, label='class 1')
axs[0].title.set_text("Naive Bayes works better")
axs[0].legend()

axs[1].hist(logistic_class_0, bins=20, alpha=0.5, label='class 0')
axs[1].hist(logistic_class_1, bins=20, alpha=0.5, label='class 1')
axs[1].title.set_text("Logistic Regression works better")
axs[1].legend()

plt.show()
```

We constructed the first dataset in such a way that for each data point we plotted we attached a Gaussian probability with different expectation and standard deviation. Because the data is based on the Gaussian we have chosen, Logistic Regression does not know how to handle it and separate the data properly. In contrast, naive bayes treats all points as independent and manages to deal with data that is distributed in the above form. For this reason the first data set will get better results with the help of naive bayes.
The second dataset, however, is based on one feature and two classes. The above feature is produced using the Logistic Regression model, so Logistic Regression knew how to handle it better, hence the above dataset will get better results with the help of this algorithm.
