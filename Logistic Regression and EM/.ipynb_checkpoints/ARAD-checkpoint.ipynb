````python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
````

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

````python


def bias_trick(X):
      return np.insert(X , obj= 0 ,values= 1 , axis= 1)

def hypothesis(X, theta):
      # print(X.shape)
      # print(theta.shape)
      return 1 / (1 + np.exp(-X.dot(theta)))

def compute_cost(X, y, theta):
    h = hypothesis(X, theta)
    return  np.sum(-y * np.log(h) - (1 - y) * np.log(1 - h)) / len(X)
  


class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    
    
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.theta = None
        

    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        # X = bias_trick(X)
        
        self.labels = set(y[:])
        np.random.seed(self.random_state)
        self.theta = np.random.random(size=X.shape[1])

        current_iter = 0
        diff_cost = self.eps + 1
        J_prev = 100
        
        while current_iter < self.n_iter and diff_cost > self.eps:
          
          sigmoid = hypothesis(X, self.theta) 
          self.theta = self.theta - (X.T.dot(sigmoid - y) * self.eta)
          
          current_J = compute_cost(X, y, self.theta)
          diff_cost = J_prev - current_J
          J_prev = current_J
          current_iter += 1
          
          if diff_cost <= self.eps:
            print(current_iter)

    def predict_instance(self, x):
      # print(f'thta: {self.theta.shape}')
      # print(f'x : {x.shape}' )
      if hypothesis(x, self.theta) < 0.5:
          return 0
      else :
        return 1
      
    def predict(self, X):
        """Return the predicted class label"""
        predict_vector = []
        X = bias_trick(X)
        for x in X:
          predict_vector.append(self.predict_instance(x))
        
        return np.array(predict_vector)
        
````

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

Shuffle the training set before you split the data to the folds.

````python
#### Your code here ####
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]


def calc_accuracy(validaton_X, validation_y, LR: LogisticRegressionGD):
    accuracy = 0 
    # validaton_X = bias_trick(validaton_X)
    for index,instance in enumerate(validaton_X):
         if LR.predict_instance(instance) == validation_y[index]:
             accuracy +=1

    return accuracy/validaton_X.shape[0]





#shuffle the data
shuffled_data = training_set.values
np.random.shuffle(shuffled_data)



#split the data to 5 folds
folded_data = np.split(shuffled_data, 5)

# fold1 , fold2, fold3, fold4, fold5 = folded_data[0] , folded_data[1], folded_data[2], folded_data[3] , folded_data[4]

params_dict = {}

for eta in etas:
    for eps in epss:
        LR = LogisticRegressionGD(eta=eta,eps=eps)
        folds_accuracy = []

        for i in range(5):
            current_validation_set = folded_data.pop(i)
            current_train_set = np.concatenate((folded_data[0] , folded_data[1], folded_data[2], folded_data[3]), axis=0)

            current_train_x, current_train_y = current_train_set[:,:-1], current_train_set[:,-1]  
            current_validation_x, current_validation_y = current_validation_set[:,:-1], current_validation_set[:,-1] 
            
            current_validation_x = bias_trick(current_validation_x)
            
            LR.fit(current_train_x, current_train_y)

            folds_accuracy.append(calc_accuracy(current_validation_x, current_validation_y, LR))
            folded_data.insert(i , current_validation_set)

        params_dict[(eta,eps)] = np.mean(folds_accuracy)


best_eta , best_eps = max(params_dict, key=params_dict.get)

print(params_dict)
print(f'best eta : {best_eta}')
print(f'best epsilon : {best_eps}')
print(f'accuracy of ({best_eta} , {best_eps} ) : {params_dict[(best_eta,best_eps)]}')






````

## Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

````python
#### Your code here ####
# calc normal pdf    
def norm_pdf(data, mu, sigma):
    exponent = -0.5 * np.square((data - mu) / sigma)
    return np.exp(exponent) / (np.sqrt(2 * np.pi) * sigma)

````

## Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

````python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1, n_iter=1000, eps=0.01):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps

        self.mu = np.array([])
        self.sigma = np.array([])
        self.weight = np.empty(k)
        self.res = np.array([])

    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        """
        self.weight.fill(1/self.k)
        split_data = np.array_split(data, self.k)
        self.mu = [data.mean() for data in split_data]
        self.sigma = [data.std() for data in split_data]


    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        self.res = np.empty([data.shape[0], self.k])
        for row in range(data.shape[0]):
          for col in range(self.k):
            self.res[row][col] = self.weight[col]*norm_pdf(data[row], self.mu[col], self.sigma[col]) 

        for i in range(data.shape[0]):
            self.res[i] /= np.sum(self.res[i])
 
           
    def maximization(self, data):
        """
        M step - updating distribution params
        """

        N = data.shape[0]

        # Calculate the new weights
        self.weight = np.sum(self.res, axis=0) / N

        # Calculate the new mean
        sum = 0 
        for j in range(self.k):
          for i in range(N):
            sum += self.res[i][j] * data[i]
          self.mu[j] = sum / (N * self.weight[j])
          sum = 0

        #Calculate the new sigma
        for j in range(self.k):
          for i in range(N):
            sum += self.res[i][j] * ((data[i] - self.mu[j]) ** 2)
          self.sigma[j] = np.sqrt(sum / (N * self.weight[j]))
          sum = 0
        



    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then expectation and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less   than eps
        or when you reach n_iter.
        """

        self.init_params(data)
    
        current_iter = 0
        diff_cost = self.eps + 1
        J_prev = 1000
        
        while current_iter < self.n_iter and diff_cost > self.eps:
          
          # Update Responsabilites
          self.expectation(data)

          #Update the rest of the parameters 
          self.maximization(data)

          current_J = self.compute_cost_EM(data)

          diff_cost = J_prev - current_J
          J_prev = current_J
          current_iter += 1
          # if diff_cost < self.eps:
          #   print(current_J)

    def compute_cost_EM(self, data):
      sum = 0
      for j in range(self.k):
        for i in range (data.shape[0]):
          sum += -np.log(self.weight[j] * norm_pdf(data[i], self.mu[j], self.sigma[j]))

      return sum

    def get_dist_params(self):
        return self.weight, self.mu, self.sigma



    
````

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


````python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1):
        self.k = k
        self.class_paramaters = {}
        self.filtered_data_by_class = {}
        self.num_of_instances = 0

    def fit(self, X, y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        self.labels = set(y[:])
        self.num_of_instances = X.shape[0]
        self.num_of_features = X.shape[1]
        
        self.filtered_data_by_class = { label : np.array(X[np.where(y[:]== label)]) for label in self.labels }
        
        em = EM(self.k)
        
        #each lable is array of N (number of featuers) tuples - each tupel is of 3 arrays  - W, MU, Sigma - each array is of k (num of gussian) values
        self.features_params_by_class = {}
        
        for label in self.labels:
          self.features_params_by_class[label] = []
          for feature_index in range(self.num_of_features):
            em.fit(self.filtered_data_by_class[label][:,feature_index])
            self.features_params_by_class[label].append(em.get_dist_params())
        
   
    
    def get_likelihood(self, x, label):
      
      res = 1 
      for i in range(self.num_of_features):
          feature_prob = 0
          
          for k in range(self.k):
            w_i = self.features_params_by_class[label][i][0][k]
            mu_i = self.features_params_by_class[label][i][1][k]
            sigma_i = self.features_params_by_class[label][i][2][k]
            
            feature_prob += w_i * norm_pdf(x[i], mu_i, sigma_i)
            
          res *= feature_prob

      return res
      
    def predict_instance(self, x):
      postrior_by_label = {}
      for label in self.labels:
        prior = self.filtered_data_by_class[label].shape[0] / self.num_of_instances
        postrior_by_label[label] = prior * self.get_likelihood(x, label)
        
      return max(postrior_by_label, key=postrior_by_label.get)
      
    def predict(self, X):
        """Return the predicted class label"""
        predict_vector = []
        for x in X:
          predict_vector.append(self.predict_instance(x))
        
        return np.array(predict_vector)
      
       
````
